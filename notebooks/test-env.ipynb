```python
import numpy as np
import cvxpy as cp
import gym
import os.path as osp
import torch
import matplotlib.pyplot as plt
import time
import torch.nn as nn

from cvxrl.utils import cvxrl_env
from cvxrl.utils.visualization_utils import draw_graph
from cvxrl.envs import planar
from cvxrl.policies import fop, actor

from cvxpylayers.torch import CvxpyLayer
```

```python
mass = 0.016
obj_mu = 1.

L = cp.Variable(9, name='l')
target_n = cp.Parameter((9,), name='target_n')
W = cp.Parameter((6,), name='w_des')
G = cp.Parameter((6, 9), name='grasp_m')
cm = np.vstack((np.eye(3), np.zeros((3, 3))))
Cm = cp.Parameter((6, 3), value=cm*mass, name='com')

f_g = np.array([0, 0, -9.81])
w_ext = -W + Cm @ f_g * mass
f = G @ L + w_ext

contact_force = L - target_n
cost = cp.sum_squares(contact_force)

constraints = []
```

```python
constraints = [cp.abs(L[2::3]) <= obj_mu * L[::3]]
constraints += [cp.abs(L[1::3]) <= obj_mu * L[::3]]
constraints.append(L[::3] >= 0)
constraints.append(f == np.zeros(f.shape))
```

```python
prob = cp.Problem(cp.Minimize(cost), constraints)
```

```python
policy = CvxpyLayer(prob, [G, W, target_n, Cm], [L])
```
